# Null Value Treatment 

```python
null=pd.DataFrame()
null['nullpercent']=(df.isnull().sum()/len(df)*100)
null.head()
```

```python
null['nullpercent'].sort_values(ascending = False)
```

```python
null_val=list(null[null['nullpercent']>50].index)
```

```python
null_val
```

```python
df=df.drop(null_val,axis=1)
df.head()
```

```python
df.shape
```

```python
uniq=pd.DataFrame()
uniq['values']=df.nunique()
uniq_val=list(uniq[uniq['values']<2].index)
uniq_val
```

```python
df = df.drop(uniq_val,axis = 1)
```

# Removing features that has future/changing info that isn't useful for our model

```python
unwanted=['url','desc','last_pymnt_d','last_pymnt_amnt','last_fico_range_high','last_fico_range_low','debt_settlement_flag',
          'last_credit_pull_d','zip_code','funded_amnt','funded_amnt_inv','sub_grade','int_rate','emp_title','issue_d',
          'total_pymnt','total_rec_int','total_rec_late_fee','recoveries', 'collection_recovery_fee','total_pymnt_inv',
          'total_rec_prncp','earliest_cr_line']
```

```python
df=df.drop(unwanted,axis=1)
df.shape
```

```python
# Making fico_average columns using fico_range_low and fico_range_high then dropping them

df['fico_average'] = (df['fico_range_high']+df['fico_range_low'])/2
df = df.drop(['fico_range_high','fico_range_low'],axis=1)
```
